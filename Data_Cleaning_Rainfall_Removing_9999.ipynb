```python
import sqlite3
import csv

# Making a New Database for the Imported csv to be stored into
original_conn = sqlite3.connect('1901_modified.db')

# The path of the CSV file to convert
csv_file_path = '/content/drive/MyDrive/rainfall_data/csv files/1901_rain.csv'

# Reading the CSV file to determine the number of columns and header row
with open(csv_file_path, 'r') as file:
    csv_reader = csv.reader(file)
    header_row = next(csv_reader)
    num_columns = len(header_row)

# Create a new database for modified data
new_db_name = '1901_modified_data.db'
modified_conn = sqlite3.connect(new_db_name)

# Create a new table with the appropriate number of columns in the modified database
table_schema = ', '.join([f'col{i}' for i in range(1, num_columns + 1)])
modified_conn.execute(f'CREATE TABLE IF NOT EXISTS csv_data ({table_schema})')

# Import CSV data into the table in the modified database
with open(csv_file_path, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    modified_conn.executemany(f'INSERT INTO csv_data VALUES ({", ".join(["?"] * num_columns)})', csv_reader)

# Delete rows with -9999 in the third column from the modified database as the rows which have -9999 values is data of outside India and of that no data is provided
modified_conn.execute('DELETE FROM csv_data WHERE col3 = "-9999"')

# Retrieve the modified data including the header row from the modified database
cursor = modified_conn.execute('SELECT * FROM csv_data')
modified_data = cursor.fetchall()

# Display the header row and modified data
print("Modified CSV file contents:")
print(header_row)
for row in modified_data:
    print(row)

cursor.close()
original_conn.close()
modified_conn.close()
```

```python

conn = sqlite3.connect('modified_data.db')

table_name = 'csv_data'

cursor = conn.execute(f'SELECT * FROM {table_name}')
data = cursor.fetchall()

csv_file_path = '/content/drive/MyDrive/rainfall_data/Main_csv/1901_cleaned.csv'

with open(csv_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

conn.close()

print('CSV file created and saved to Google Drive:', csv_file_path)

```
